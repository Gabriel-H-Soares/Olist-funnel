### Concepts of Metrics and the Confusion Matrix

- ROC-AUC: Measures the model’s ability to distinguish between classes (churn vs. non-churn) across all possible thresholds. The closer to 1, the better the separation.
- Precision: Of all customers predicted as churn, how many actually churned.
- Recall: Of all customers who actually churned, how many the model correctly identified.
- F1-score: The harmonic mean of precision and recall. It is useful when you want to balance both metrics.
- Confusion Matrix: A table that summarizes the model’s correct and incorrect predictions.
    - TN (True Negative): Predicted non-churn and it was actually non-churn.
    - FP (False Positive): Predicted churn, but it was actually non-churn.
    - FN (False Negative): Predicted non-churn, but it was actually churn.
    - TP (True Positive): Predicted churn and it was actually churn.

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import (accuracy_score, roc_auc_score, precision_score, recall_score, f1_score, confusion_matrix)
import plotly.express as px

DF_Model = DF_IBM.copy()

X_Features = DF_Model.drop(columns=['customerID', 'gender', 'MultipleLines', 'PhoneService', 'Churn'])
Y_Target = DF_Model['Churn']

x_train, x_test, y_train, y_test = train_test_split(X_Features, Y_Target, test_size=0.2, random_state=42, stratify=Y_Target)

Scaler = StandardScaler()
X_Features_scaled = Scaler.fit_transform(X_Features)


Models = {
    'Random Forest': RandomForestClassifier(n_estimators=1000, max_depth=3 , min_samples_leaf=0.022713045260637697, 
                                            min_samples_split=0.060777644831336686, max_features=0.4764433036390532, max_samples=0.8137950277696371, ccp_alpha=0.0006766355550890428, 
                                            criterion='log_loss', class_weight='balanced_subsample', n_jobs=-1, random_state=42),
    'Logistic Regression': Pipeline(steps=[('scaler', StandardScaler()), ('LR_classifier', LogisticRegression(max_iter=3000, C=0.7, class_weight='balanced', random_state=42))]),
    'XGBoost': XGBClassifier(n_estimators=1800, max_depth=7, min_child_weight=12, subsample=0.518429747029322, colsample_bytree=0.8004792388244336, gamma=3.8352981864500997, reg_lambda=19.790179118425787, reg_alpha=5.4376274935084304e-05, learning_rate=0.01819158679868869, 
                             eval_metric='auc', early_stopping_rounds=120, n_jobs=-1, random_state=42)
} 

# For early stopping in XGBoost
x, x_val, y, y_val = train_test_split(
    x_train,
    y_train,
    test_size=0.2,
    random_state=42,
    stratify=y_train
)

for name, model in Models.items():

    # Training the model
    if name == 'XGBoost':
        model.fit(x, y, eval_set=[(x_val, y_val)], verbose=False)
    else:
        model.fit(x_train, y_train)

    # Predictions
    y_pred = model.predict(x_test)
    y_prob = model.predict_proba(x_test)[:, 1]

    # Metrics  
    Accuracy = accuracy_score(y_test, y_pred)
    ROC_AUC = roc_auc_score(y_test, y_prob)
    Precision = precision_score(y_test, y_pred)
    Recall = recall_score(y_test, y_pred)
    F1 = f1_score(y_test, y_pred)

    # Results
    print(f'Test Results: {name}')
    print(f'Accuracy : {Accuracy:.4f}')
    print(f'ROC-AUC : {ROC_AUC:.4f}')
    print(f'Precision: {Precision:.4f}')
    print(f'Recall   : {Recall:.4f}')
    print(f'F1-score : {F1:.4f}')

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    tn, fp, fn, tp = cm.ravel()
    print(f'\nConfusion Matrix - {name}')
    print(f'TP: {tp}, FN: {fn}, FP: {fp}, TN: {tn}')
    labels = ['Nao Churn', 'Churn']
    fig = px.imshow(cm, x=labels, y=labels, text_auto=True, color_continuous_scale='Blues', labels=dict(x="Predicted", y="Actual", color="Count"))
    fig.update_layout(title=f'Confusion Matrix - {name}', xaxis_title='Predicted', yaxis_title='Actual', width=500, height=400)
    fig.show()

    if name == 'Random Forest':
        Model_RF = model
        ROC_AUC_RF = ROC_AUC
    elif name == 'Logistic Regression':
        Model_LR = model
        ROC_AUC_LR = ROC_AUC
    else:
        Model_XGB = model
        ROC_AUC_XGB = ROC_AUC

# Result by customerID in Test dataset
Prob_test_RF = Model_RF.predict_proba(x_test)[:,1]
Pred_Test_RF = Model_RF.predict(x_test)
DF_Predictions_Test_RF = pd.DataFrame({
    'customerID': DF_Model.loc[x_test.index, 'customerID'],
    'churn_prob': Prob_test_RF,
    'predict_churn': np.where(Pred_Test_RF == 1, 'Positive', 'Negative')
}).sort_values(by='churn_prob', ascending=False)

print('\nRandom Forest: Top 5 customerID test dataset with highest churn risk:')
display(DF_Predictions_Test_RF.head(5))

Prob_test_LR = Model_LR.predict_proba(x_test)[:,1]
Pred_Test_LR = Model_LR.predict(x_test)
DF_Predictions_Test_LR = pd.DataFrame({
    'customerID': DF_Model.loc[x_test.index, 'customerID'],
    'churn_prob': Prob_test_LR,
    'predict_churn': np.where(Pred_Test_LR == 1, 'Positive', 'Negative')
}).sort_values(by='churn_prob', ascending=False)

print('\nLogistic Regression: Top 5 customerID test dataset with highest churn risk:')
display(DF_Predictions_Test_LR.head(5))

Prob_test_XGB = Model_XGB.predict_proba(x_test)[:,1]
Pred_Test_XGB = Model_XGB.predict(x_test)
DF_Predictions_Test_XGB = pd.DataFrame({
    'customerID': DF_Model.loc[x_test.index, 'customerID'],
    'churn_prob': Prob_test_XGB,
    'predict_churn': np.where(Pred_Test_XGB == 1, 'Positive', 'Negative')
}).sort_values(by='churn_prob', ascending=False)

print('\nXGBoost: Top 5 customerID test dataset with highest churn risk:')
display(DF_Predictions_Test_XGB.head(5))
```

```python
# Cross-validation with 5 folds
from sklearn.model_selection import StratifiedKFold, cross_validate, train_test_split
from sklearn.base import clone

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

scoring = {
    'accuracy': 'accuracy',
    'roc_auc': 'roc_auc',
    'precision': 'precision',
    'recall': 'recall',
    'f1': 'f1'
}

for name, model in Models.items():
    print(f'\nCross-validation Results: {name}')
    print(f'{"Metric":<10} {"Mean":>8} {"Std":>8}')

    if name != 'XGBoost':
        model_cv = clone(model)
        cv_results = cross_validate(
            model_cv,
            X_Features,
            Y_Target,
            cv=cv,
            scoring=scoring,
            n_jobs=-1,
            return_train_score=False
        )

        for metric in scoring.keys():
            metric_values = cv_results[f'test_{metric}']
            print(f'{metric.capitalize():<10} {metric_values.mean():>8.4f} {metric_values.std():>8.4f}')
    else:
        # Manual CV so each fold can use an internal eval_set for early stopping
        fold_scores = {k: [] for k in scoring.keys()}

        for fold_idx, (train_idx, test_idx) in enumerate(cv.split(X_Features, Y_Target), start=1):
            X_tr = X_Features.iloc[train_idx]
            y_tr = Y_Target.iloc[train_idx]
            X_te = X_Features.iloc[test_idx]
            y_te = Y_Target.iloc[test_idx]

            X_fit, X_es, y_fit, y_es = train_test_split(
                X_tr,
                y_tr,
                test_size=0.2,
                random_state=42 + fold_idx,
                stratify=y_tr
            )

            model_cv = clone(model)
            model_cv.fit(X_fit, y_fit, eval_set=[(X_es, y_es)], verbose=False)

            y_pred = model_cv.predict(X_te)
            y_prob = model_cv.predict_proba(X_te)[:, 1]

            fold_scores['accuracy'].append(accuracy_score(y_te, y_pred))
            fold_scores['roc_auc'].append(roc_auc_score(y_te, y_prob))
            fold_scores['precision'].append(precision_score(y_te, y_pred))
            fold_scores['recall'].append(recall_score(y_te, y_pred))
            fold_scores['f1'].append(f1_score(y_te, y_pred))

        for metric in scoring.keys():
            metric_values = np.array(fold_scores[metric])
            print(f'{metric.capitalize():<10} {metric_values.mean():>8.4f} {metric_values.std():>8.4f}')
```

```python
# Consolidated ranking with ROC-AUC + Recall (Test and Cross-Validation)
from sklearn.model_selection import cross_validate, train_test_split
from sklearn.base import clone

comparison_rows = []

for name, model in Models.items():
    Model = clone(model)

    # Test metrics
    if name == 'XGBoost':
        # Use internal validation split to activate early stopping in holdout training
        Model.fit(x, y, eval_set=[(x_val, y_val)], verbose=False)
    else:
        Model.fit(x_train, y_train)

    y_test_prob = Model.predict_proba(x_test)[:, 1]
    y_test_pred = (y_test_prob >= 0.5).astype(int)
    test_roc_auc = roc_auc_score(y_test, y_test_prob)
    test_recall = recall_score(y_test, y_test_pred)

    # Cross-validation metrics
    if name != 'XGBoost':
        cv_scores = cross_validate(
            Model,
            X_Features,
            Y_Target,
            cv=cv,
            scoring={'roc_auc': 'roc_auc', 'recall': 'recall'},
            n_jobs=-1,
            return_train_score=False
        )
        cv_roc_auc = cv_scores['test_roc_auc'].mean()
        cv_recall = cv_scores['test_recall'].mean()
    else:
        # Manual CV so each fold can use an internal eval_set for early stopping
        fold_auc = []
        fold_recall = []

        for fold_idx, (train_idx, test_idx) in enumerate(cv.split(X_Features, Y_Target), start=1):
            X_tr = X_Features.iloc[train_idx]
            y_tr = Y_Target.iloc[train_idx]
            X_te = X_Features.iloc[test_idx]
            y_te = Y_Target.iloc[test_idx]

            X_fit, X_es, y_fit, y_es = train_test_split(
                X_tr,
                y_tr,
                test_size=0.2,
                random_state=42 + fold_idx,
                stratify=y_tr
            )

            model_cv = clone(model)
            model_cv.fit(X_fit, y_fit, eval_set=[(X_es, y_es)], verbose=False)

            y_cv_prob = model_cv.predict_proba(X_te)[:, 1]
            y_cv_pred = model_cv.predict(X_te)

            fold_auc.append(roc_auc_score(y_te, y_cv_prob))
            fold_recall.append(recall_score(y_te, y_cv_pred))

        cv_roc_auc = float(np.mean(fold_auc))
        cv_recall = float(np.mean(fold_recall))

    # Composite score giving same weight to Test and CV behavior
    final_score = np.mean([test_roc_auc, test_recall, cv_roc_auc, cv_recall])

    comparison_rows.append({
        'Model': name,
        'Test_ROC_AUC': test_roc_auc,
        'Test_Recall': test_recall,
        'CV_ROC_AUC': cv_roc_auc,
        'CV_Recall': cv_recall,
        'Final_Score': final_score
    })

DF_Comparison = pd.DataFrame(comparison_rows).sort_values('Final_Score', ascending=False).reset_index(drop=True)

display(DF_Comparison.style.format({
    'Test_ROC_AUC': '{:.4f}',
    'Test_Recall': '{:.4f}',
    'CV_ROC_AUC': '{:.4f}',
    'CV_Recall': '{:.4f}',
    'Final_Score': '{:.4f}'
}))

best_model = DF_Comparison.loc[0, 'Model']
print(f"Best model by ROC-AUC + Recall (Test + CV): {best_model}")
```

```python
# Overfitting diagnostic with statistical classification
from sklearn.base import clone
from sklearn.model_selection import train_test_split

def mean_ci95(values):
    arr = np.asarray(values, dtype=float)
    n = len(arr)
    mean_val = float(np.mean(arr))
    if n < 2:
        return mean_val, np.nan, np.nan
    se = float(np.std(arr, ddof=1) / np.sqrt(n))
    z = 1.96
    return mean_val, mean_val - z * se, mean_val + z * se

diagnostic_rows = []

for name, model in Models.items():
    model_diag = clone(model)

    # Holdout fit (train/test)
    if name == 'XGBoost':
        X_fit_hold, X_es_hold, y_fit_hold, y_es_hold = train_test_split(
            x_train,
            y_train,
            test_size=0.2,
            random_state=42,
            stratify=y_train
        )
        model_diag.fit(X_fit_hold, y_fit_hold, eval_set=[(X_es_hold, y_es_hold)], verbose=False)
    else:
        model_diag.fit(x_train, y_train)

    # Holdout metrics
    y_train_prob = model_diag.predict_proba(x_train)[:, 1]
    y_train_pred = (y_train_prob >= 0.5).astype(int)
    train_roc_auc = roc_auc_score(y_train, y_train_prob)
    train_recall = recall_score(y_train, y_train_pred)

    y_test_prob = model_diag.predict_proba(x_test)[:, 1]
    y_test_pred = (y_test_prob >= 0.5).astype(int)
    test_roc_auc = roc_auc_score(y_test, y_test_prob)
    test_recall = recall_score(y_test, y_test_pred)

    # Manual CV fold-by-fold (supports early stopping for XGBoost)
    fold_train_auc = []
    fold_test_auc = []
    fold_train_recall = []
    fold_test_recall = []

    for fold_idx, (train_idx, test_idx) in enumerate(cv.split(X_Features, Y_Target), start=1):
        X_tr = X_Features.iloc[train_idx]
        y_tr = Y_Target.iloc[train_idx]
        X_te = X_Features.iloc[test_idx]
        y_te = Y_Target.iloc[test_idx]

        model_fold = clone(model)

        if name == 'XGBoost':
            X_fit, X_es, y_fit, y_es = train_test_split(
                X_tr,
                y_tr,
                test_size=0.2,
                random_state=42 + fold_idx,
                stratify=y_tr
            )
            model_fold.fit(X_fit, y_fit, eval_set=[(X_es, y_es)], verbose=False)
        else:
            model_fold.fit(X_tr, y_tr)

        y_tr_prob = model_fold.predict_proba(X_tr)[:, 1]
        y_tr_pred = (y_tr_prob >= 0.5).astype(int)
        y_te_prob = model_fold.predict_proba(X_te)[:, 1]
        y_te_pred = (y_te_prob >= 0.5).astype(int)

        fold_train_auc.append(roc_auc_score(y_tr, y_tr_prob))
        fold_test_auc.append(roc_auc_score(y_te, y_te_prob))
        fold_train_recall.append(recall_score(y_tr, y_tr_pred))
        fold_test_recall.append(recall_score(y_te, y_te_pred))

    gap_auc_folds = np.asarray(fold_train_auc) - np.asarray(fold_test_auc)
    gap_recall_folds = np.asarray(fold_train_recall) - np.asarray(fold_test_recall)

    gap_auc_mean, gap_auc_ci_low, gap_auc_ci_high = mean_ci95(gap_auc_folds)
    gap_recall_mean, gap_recall_ci_low, gap_recall_ci_high = mean_ci95(gap_recall_folds)

    # Statistical classification using 95% CI around fold-level gaps
    auc_overfit = gap_auc_ci_low > 0
    recall_overfit = gap_recall_ci_low > 0
    auc_underfit = gap_auc_ci_high < 0
    recall_underfit = gap_recall_ci_high < 0

    if auc_overfit and recall_overfit:
        overfit_level = 'Overfitting confirmed (AUC + Recall, 95% CI)'
    elif auc_overfit or recall_overfit:
        overfit_level = 'Overfitting signal in one metric (95% CI)'
    elif auc_underfit and recall_underfit:
        overfit_level = 'Possible underfitting (both CIs below zero)'
    else:
        overfit_level = 'No statistical evidence of overfitting (95% CI)'

    diagnostic_rows.append({
        'Model': name,
        'Train_ROC_AUC': train_roc_auc,
        'Test_ROC_AUC': test_roc_auc,
        'Train_Recall': train_recall,
        'Test_Recall': test_recall,
        'Gap_AUC_Mean': gap_auc_mean,
        'Gap_AUC_CI95_Low': gap_auc_ci_low,
        'Gap_AUC_CI95_High': gap_auc_ci_high,
        'Gap_Recall_Mean': gap_recall_mean,
        'Gap_Recall_CI95_Low': gap_recall_ci_low,
        'Gap_Recall_CI95_High': gap_recall_ci_high,
        'Overfitting_Level': overfit_level
    })

DF_Overfit = pd.DataFrame(diagnostic_rows)

severity_rank = {
    'Overfitting confirmed (AUC + Recall, 95% CI)': 0,
    'Overfitting signal in one metric (95% CI)': 1,
    'No statistical evidence of overfitting (95% CI)': 2,
    'Possible underfitting (both CIs below zero)': 3
}

DF_Overfit['Severity_Rank'] = DF_Overfit['Overfitting_Level'].map(severity_rank)
DF_Overfit = DF_Overfit.sort_values(
    by=['Severity_Rank', 'Gap_AUC_Mean', 'Gap_Recall_Mean'],
    ascending=[True, False, False]
).drop(columns=['Severity_Rank']).reset_index(drop=True)

display(DF_Overfit.style.format({
    'Train_ROC_AUC': '{:.4f}',
    'Test_ROC_AUC': '{:.4f}',
    'Train_Recall': '{:.4f}',
    'Test_Recall': '{:.4f}',
    'Gap_AUC_Mean': '{:.4f}',
    'Gap_AUC_CI95_Low': '{:.4f}',
    'Gap_AUC_CI95_High': '{:.4f}',
    'Gap_Recall_Mean': '{:.4f}',
    'Gap_Recall_CI95_Low': '{:.4f}',
    'Gap_Recall_CI95_High': '{:.4f}'
}))

print('\nStatistical overfitting summary (95% CI):')
for _, row in DF_Overfit.iterrows():
    print(
        f"- {row['Model']}: {row['Overfitting_Level']} "
        f"| AUC gap CI95=[{row['Gap_AUC_CI95_Low']:.4f}, {row['Gap_AUC_CI95_High']:.4f}] "
        f"| Recall gap CI95=[{row['Gap_Recall_CI95_Low']:.4f}, {row['Gap_Recall_CI95_High']:.4f}]"
    )
```

```python
# Optuna tuning for all models focused on overfitting mitigation
import optuna

cv_opt = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

def evaluate_model_cv(model_name, estimator):
    # Returns metrics needed to quantify overfitting and generalization
    if model_name != 'XGBoost':
        cv_res = cross_validate(
            estimator,
            X_Features,
            Y_Target,
            cv=cv_opt,
            scoring={'roc_auc': 'roc_auc', 'recall': 'recall'},
            return_train_score=True,
            n_jobs=-1,
            error_score='raise'
        )
        cv_auc = float(np.mean(cv_res['test_roc_auc']))
        cv_recall = float(np.mean(cv_res['test_recall']))
        train_auc = float(np.mean(cv_res['train_roc_auc']))
        train_recall = float(np.mean(cv_res['train_recall']))
        auc_std = float(np.std(cv_res['test_roc_auc'], ddof=1))
        recall_std = float(np.std(cv_res['test_recall'], ddof=1))
        return cv_auc, cv_recall, train_auc, train_recall, auc_std, recall_std

    # Manual CV for XGBoost so each fold can use early stopping with internal eval_set
    fold_train_auc, fold_test_auc = [], []
    fold_train_recall, fold_test_recall = [], []

    for fold_idx, (train_idx, test_idx) in enumerate(cv_opt.split(X_Features, Y_Target), start=1):
        X_tr = X_Features.iloc[train_idx]
        y_tr = Y_Target.iloc[train_idx]
        X_te = X_Features.iloc[test_idx]
        y_te = Y_Target.iloc[test_idx]

        X_fit, X_es, y_fit, y_es = train_test_split(
            X_tr,
            y_tr,
            test_size=0.2,
            random_state=42 + fold_idx,
            stratify=y_tr
        )

        model_fold = estimator
        model_fold.fit(X_fit, y_fit, eval_set=[(X_es, y_es)], verbose=False)

        y_tr_prob = model_fold.predict_proba(X_tr)[:, 1]
        y_tr_pred = (y_tr_prob >= 0.5).astype(int)
        y_te_prob = model_fold.predict_proba(X_te)[:, 1]
        y_te_pred = (y_te_prob >= 0.5).astype(int)

        fold_train_auc.append(roc_auc_score(y_tr, y_tr_prob))
        fold_test_auc.append(roc_auc_score(y_te, y_te_prob))
        fold_train_recall.append(recall_score(y_tr, y_tr_pred))
        fold_test_recall.append(recall_score(y_te, y_te_pred))

    cv_auc = float(np.mean(fold_test_auc))
    cv_recall = float(np.mean(fold_test_recall))
    train_auc = float(np.mean(fold_train_auc))
    train_recall = float(np.mean(fold_train_recall))
    auc_std = float(np.std(fold_test_auc, ddof=1))
    recall_std = float(np.std(fold_test_recall, ddof=1))
    return cv_auc, cv_recall, train_auc, train_recall, auc_std, recall_std

def build_estimator(model_name, trial):
    if model_name == 'Random Forest':
        params = {
            'n_estimators': trial.suggest_int('n_estimators', 200, 1200, step=100),
            'max_depth': trial.suggest_int('max_depth', 3, 12),
            'min_samples_leaf': trial.suggest_float('min_samples_leaf', 0.005, 0.05),
            'min_samples_split': trial.suggest_float('min_samples_split', 0.01, 0.15),
            'max_features': trial.suggest_float('max_features', 0.2, 0.9),
            'max_samples': trial.suggest_float('max_samples', 0.5, 0.95),
            'ccp_alpha': trial.suggest_float('ccp_alpha', 1e-6, 3e-3, log=True),
            'criterion': trial.suggest_categorical('criterion', ['gini', 'entropy', 'log_loss']),
            'class_weight': trial.suggest_categorical('class_weight', ['balanced', 'balanced_subsample', None]),
            'bootstrap': True,
            'n_jobs': -1,
            'random_state': 42
        }
        return RandomForestClassifier(**params)

    if model_name == 'Logistic Regression':
        c_value = trial.suggest_float('C', 1e-3, 30.0, log=True)
        class_weight = trial.suggest_categorical('class_weight', ['balanced', None])
        return Pipeline(steps=[
            ('scaler', StandardScaler()),
            ('LR_classifier', LogisticRegression(
                C=c_value,
                max_iter=4000,
                class_weight=class_weight,
                solver='lbfgs',
                random_state=42
            ))
        ])

    # XGBoost
    from xgboost import XGBClassifier
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 400, 3000, step=200),
        'max_depth': trial.suggest_int('max_depth', 2, 8),
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 12),
        'subsample': trial.suggest_float('subsample', 0.5, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 1.0),
        'gamma': trial.suggest_float('gamma', 0.0, 5.0),
        'reg_lambda': trial.suggest_float('reg_lambda', 0.5, 20.0, log=True),
        'reg_alpha': trial.suggest_float('reg_alpha', 1e-6, 5.0, log=True),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.2, log=True),
        'eval_metric': 'auc',
        'early_stopping_rounds': 120,
        'n_jobs': -1,
        'random_state': 42
    }
    return XGBClassifier(**params)

def make_objective(model_name):
    def objective(trial):
        estimator = build_estimator(model_name, trial)
        cv_auc, cv_recall, train_auc, train_recall, auc_std, recall_std = evaluate_model_cv(model_name, estimator)

        gap_auc = train_auc - cv_auc
        gap_recall = train_recall - cv_recall

        # Primary target: minimize overfitting and instability
        overfit_penalty = max(gap_auc, 0.0) + max(gap_recall, 0.0)
        instability_penalty = auc_std + recall_std

        # Secondary: keep minimum predictive quality
        weak_perf_penalty = max(0.60 - cv_auc, 0.0) + max(0.55 - cv_recall, 0.0)

        objective_value = overfit_penalty + 0.5 * instability_penalty + 2.0 * weak_perf_penalty

        trial.set_user_attr('cv_auc', cv_auc)
        trial.set_user_attr('cv_recall', cv_recall)
        trial.set_user_attr('train_auc', train_auc)
        trial.set_user_attr('train_recall', train_recall)
        trial.set_user_attr('gap_auc', gap_auc)
        trial.set_user_attr('gap_recall', gap_recall)
        trial.set_user_attr('auc_std', auc_std)
        trial.set_user_attr('recall_std', recall_std)
        trial.set_user_attr('overfit_penalty', overfit_penalty)
        trial.set_user_attr('instability_penalty', instability_penalty)
        trial.set_user_attr('weak_perf_penalty', weak_perf_penalty)
        return objective_value
    return objective

optuna.logging.set_verbosity(optuna.logging.WARNING)

models_to_tune = ['Random Forest', 'Logistic Regression', 'XGBoost']
n_trials_by_model = {'Random Forest': 50, 'Logistic Regression': 35, 'XGBoost': 45}

best_estimators = {}
summary_rows = []
all_studies = {}

for model_name in models_to_tune:
    print(f'\n===== Optuna tuning: {model_name} =====')
    sampler = optuna.samplers.TPESampler(seed=42)
    study = optuna.create_study(
        study_name=f'optuna_{model_name.lower().replace(" ", "_")}',
        direction='minimize',
        sampler=sampler
    )
    study.optimize(make_objective(model_name), n_trials=n_trials_by_model[model_name], show_progress_bar=True)

    all_studies[model_name] = study
    best_trial = study.best_trial
    best_estimator = build_estimator(model_name, best_trial)
    best_estimators[model_name] = best_estimator

    summary_rows.append({
        'Model': model_name,
        'Objective_Min': best_trial.value,
        'CV_AUC': best_trial.user_attrs['cv_auc'],
        'CV_Recall': best_trial.user_attrs['cv_recall'],
        'Train_AUC': best_trial.user_attrs['train_auc'],
        'Train_Recall': best_trial.user_attrs['train_recall'],
        'Gap_AUC': best_trial.user_attrs['gap_auc'],
        'Gap_Recall': best_trial.user_attrs['gap_recall'],
        'AUC_STD': best_trial.user_attrs['auc_std'],
        'Recall_STD': best_trial.user_attrs['recall_std'],
        'Overfit_Penalty': best_trial.user_attrs['overfit_penalty'],
        'Instability_Penalty': best_trial.user_attrs['instability_penalty'],
        'WeakPerf_Penalty': best_trial.user_attrs['weak_perf_penalty'],
        'Best_Params': best_trial.params
    })

DF_Optuna_AllModels = pd.DataFrame(summary_rows).sort_values('Objective_Min', ascending=True).reset_index(drop=True)

display(DF_Optuna_AllModels[['Model', 'Objective_Min', 'CV_AUC', 'CV_Recall', 'Gap_AUC', 'Gap_Recall', 'AUC_STD', 'Recall_STD']].style.format({
    'Objective_Min': '{:.5f}',
    'CV_AUC': '{:.4f}',
    'CV_Recall': '{:.4f}',
    'Gap_AUC': '{:.4f}',
    'Gap_Recall': '{:.4f}',
    'AUC_STD': '{:.4f}',
    'Recall_STD': '{:.4f}'
}))

print('\nBest parameters per model (focus: overfitting mitigation):')
for model_name in models_to_tune:
    print(f'\n{model_name}:')
    print(all_studies[model_name].best_trial.params)

# Optional: use tuned estimators in downstream evaluation blocks
Models_Tuned = best_estimators.copy()
```
